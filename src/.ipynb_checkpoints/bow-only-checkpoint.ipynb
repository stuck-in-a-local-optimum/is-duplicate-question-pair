```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading train data
df = pd.read_csv('../data/train.csv')
df.shape
```

```python
df.head()
```

```python
#large-samples so taking random 30k samples only as of now
new_df = df.sample(30000)
```

```python
new_df.isnull().sum()
```

```python
questn_df = new_df[['question1', 'question2']]
questn_df.head()
```

```python
#representing questions text to bag-of-words
from sklearn.feature_extraction.text import CountVectorizer
#marge texts
#all_questions first 30k question1 and rest 30k question2
all_questions = list(questn_df['question1']) + list(questn_df['question2'])

count_vectorizer = CountVectorizer(max_features=3000)

#get the bow-representation if each questions in question1 and question2
q1_arr, q2_arr = np.vsplit(count_vectorizer.fit_transform(all_questions).toarray(), 2)
```

```python
#converting these bow to dataframe
temp_df1 = pd.DataFrame(q1_arr, index = questn_df.index)
temp_df2 = pd.DataFrame(q2_arr, index = questn_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```

We have 30k samples and each sample have <br>
3000 bow features for question1 and 3000 for question2,<br>total 6000 features of each sample

```python
temp_df
```

```python
#adding the label column in the temp_df
temp_df['is_duplicate'] = new_df['is_duplicate']
```

```python
temp_df.head()
```

```python
#splitting the dataset in train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(temp_df.iloc[:,0:-1].values,temp_df.iloc[:,-1].values,test_size=0.2,random_state=1)
```

```python
#checking performance on RF
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test, y_pred)
```
